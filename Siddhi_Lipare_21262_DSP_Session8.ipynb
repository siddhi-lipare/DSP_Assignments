Q5:   
(a)

```python
data6 = {
    'Sr.no': [0, 1, 2, 3, 4, 5],
    'Name': ['Ashwini', 'Aditi', 'Siddhi', 'Kshiti', 'Anagha', 'Sharry'],
    'Number': [101.1, 202.2, 303.3, 404.4, 505.5, 606.6],
    'Vegetable': ['Cauliflower', 'Tomato', 'Potato', 'Okra', 'Peanuts', 'Beans'],
    'Fruit': ['Apple', 'Banana', 'Cherry', 'Orange', 'Pear', 'Fig']
}

# Creating a DataFrame from the above dictionary
my_df = pd.DataFrame(data6)

# Creating a new column using a dictionary
new_column = {0: 'Hibiscus', 1: 'Rose', 2: 'Daisy', 3: 'Tulip', 4: 'Sunflower', 5: 'Marigold'}
my_df['Flower'] = my_df['Sr.no'].map(new_column)

my_df
```

(b)

```python
# Iterate over rows using iterrows()
for index, row in my_df.iterrows():
    print(f"Row {index + 1}:")
    print(f"Sr.no: {row['Sr.no']}")
    print(f"Name: {row['Name']}")
    print(f"Number: {row['Number']}")
    print(f"Vegetable: {row['Vegetable']}")
    print(f"Fruit: {row['Fruit']}")
    print(f"Flower: {row['Flower']}")
    print("---------------------------")
```

(c)

```python
# Get column headers
column_headers = my_df.columns
column_headers
```

(d)

```python
# Select 3 rows and 3 columns using iloc[] (by integer position)
selected_data_iloc = my_df.iloc[:3, :3]

# Select 3 rows and 3 columns using loc[] (by label)
selected_data_loc = my_df.loc[:2, ['Sr.no', 'Name', 'Number']]

print("Using iloc[]:\n", selected_data_iloc)

print("\nUsing loc[]:\n", selected_data_loc)
```

(e)

```python
# Select the last row, last column element using iat[]
last_element_iat = my_df.iat[-1, -1]

# Select the last row, last column element using at[]
last_element_at = my_df.at[len(my_df) - 1, 'Flower']

print("Using iat[]:", last_element_iat)
print("Using at[]:", last_element_at)
```

Q6:       
(a)

```python
# Create dataframes
my_data1 = {'Siddhi': [10, 20], 'Sharry': [30, 40]}
my_data2 = {'Siddhi': [50, 60], 'Sharry': [70, 80]}
my_data3 = {'Avni': [15, 20], 'Aditi': [35, 40]}
my_data4 = {'Avni': [65, 75], 'Aditi': [80, 90]}

mydf1 = pd.DataFrame(my_data1)
mydf2 = pd.DataFrame(my_data2)
mydf3 = pd.DataFrame(my_data3)
mydf4 = pd.DataFrame(my_data4)

# Concatenate all dataframes into a new dataframe
concatenated_dfs = pd.concat([mydf1, mydf2, mydf3, mydf4])

print("Concatenated DataFrame:")
concatenated_dfs
```

(b)

```python
# Add a key to each dataframe
mydf1['Key'] = 'mydf1'
mydf2['Key'] = 'mydf2'
mydf3['Key'] = 'mydf3'
mydf4['Key'] = 'mydf4'

# Concatenate the dataframes into a new dataframe with a key
concatenated_df_with_key = pd.concat([mydf1, mydf2, mydf3, mydf4])
print("\nConcatenated DataFrame with Key:")
concatenated_df_with_key
```

(c)

```python
# Create a series
new_series = pd.Series([100, 120], name='Anagha')

# Append the series to df1
appenddf1 = mydf1.join(new_series)

print("\nDataFrame mydf1 with Appended Series:")
appenddf1
```

(d)

```python
inner_join_mydf = pd.merge(mydf1, mydf2, on='Sharry', how='inner')
outer_join_mydf = pd.merge(mydf1, mydf2, on='Siddhi', how='outer')
left_join_mydf = pd.merge(mydf1, mydf2, on='Siddhi', how='left')
right_join_mydf = pd.merge(mydf1, mydf2, on='Siddhi', how='right')

print("\nInner Join:")
print(inner_join_mydf)
print("\nOuter Join:")
print(outer_join_mydf)
print("\nLeft Join:")
print(left_join_mydf)
print("\nRight Join:")
print(right_join_mydf)
```
